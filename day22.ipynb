```python
top10_value = df.sort_values(
    "Total Traded Value", ascending=False
).head(10)

plt.bar(top10_value["Symbol"], top10_value["Total Traded Value"])
plt.title("Top 10 Securities by Total Traded Value")
plt.xticks(rotation=45, ha='right') 

plt.xlabel("Symbol")
plt.ylabel("Total Traded Value")
plt.show()
```

```python
top5_mcap = df.sort_values(
    "Market Capitalization", ascending=False
).head(5)
plt.pie(
    top5_mcap["Market Capitalization"],
    labels=top5_mcap["Symbol"],
    autopct= '%1.1f%%'
)
plt.title("Market Capitalization Share (Top 5)")
plt.show()
```
